```python
#Adding a new column that determines the gender and age of the person in the thumbnail
import cv2
from pyagender import PyAgender
import urllib.request as urllib

def detectThumbnail(data):
    age_n_gender_Arr = PyAgender()
    detectionList = []
    for d in data:
        url = "https://img.youtube.com/vi/"+str(d)+"0.jpg"
        try:
            img = np.asarray(bytearray(urllib.urlopen(url).read()), dtype="uint8")
            imgOutputArray = age_n_gender_Arr.detect_genders_ages(cv2.imdecode(img, cv2.IMREAD_COLOR))
            detectionList.extend([round(imgOutputArray[0]["gender"]),round(imgOutputArray[0]["age"])])
        except:
            detectionList.extend([])
    return detectionList
imgDf = pd.Series(df["videoId"])
imgDf.apply(detectThumbnail)
```
