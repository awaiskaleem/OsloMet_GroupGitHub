```python
df["Title_Length"] = df["Title"].apply(len)
df["Title_UpLow"] = df['Title'].apply(lambda x: True if (x.isupper()) | (x.islower()) else False)
```

```python
#Creating new columns for the Trending Weekday and Publish Weekday
import datetime
weekdays = {0:"Monday",
           1: "Tuesday",
           2:"Wednesday",
           3:"Thursday",
           4: "Friday",
           5:"Saturday",
           6: "Sunday"}
df["publish_weekday"] = df["PublishedAt"].apply(lambda x : datetime.datetime(int(str(x)[:4]),int(str(x)[5:7]),int(str(x)[8:10])).weekday())
df["publish_weekday"].replace(weekdays, inplace=True) 
```

```python
def dayTimeMap(hour):
    hour = int(hour[11:13])
    if (hour <= 11):
        return 'Morning'
    elif (hour > 11) and (hour <= 14 ):
        return 'Noon'
    elif (hour > 14) and (hour <= 17):
        return'Afternoon'
    elif (hour > 17) and (hour <= 21) :
        return 'Evening'
    elif (hour > 21) and (hour <= 24):
        return'Night'

df["PublishDayTime"] = df["PublishedAt"].apply(dayTimeMap)
```
